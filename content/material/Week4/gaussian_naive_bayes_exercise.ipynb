## Step 4: Mathematical Foundation

Before implementing, let's understand the math behind Gaussian Naive Bayes:

### Bayes' Theorem
$$P(C_k | \mathbf{x}) = \frac{P(\mathbf{x} | C_k) P(C_k)}{P(\mathbf{x})}$$

### Naive Bayes Assumption
Features are conditionally independent:
$$P(\mathbf{x} | C_k) = \prod_{i=1}^{n} P(x_i | C_k)$$

### Gaussian Distribution
For continuous features, we assume Gaussian distribution:
$$P(x_i | C_k) = \frac{1}{\sqrt{2\pi\sigma_{ik}^2}} \exp\left(-\frac{(x_i - \mu_{ik})^2}{2\sigma_{ik}^2}\right)$$

Where:
- $\mu_{ik}$ = mean of feature $i$ for class $k$
- $\sigma_{ik}^2$ = variance of feature $i$ for class $k$
